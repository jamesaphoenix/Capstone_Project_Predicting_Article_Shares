```python
def correlation_heat_map(df):
    corrs = df.corr()
    corrs = df.corr()[(df.corr() < 1.0) & (df.corr() > 0.25) | (df.corr() < -0.25)].dropna(how='all')
    # Set the default matplotlib figure size:
    fig, ax = plt.subplots(figsize=(22, 10))

    # # Plot the heatmap with seaborn.
    ax = sns.heatmap(corrs, annot=True, cmap='coolwarm', linewidths=2)
    label = ax.set_title('Investigating The Correlations Between Article Variables', fontsize = 24, pad=30)
```

## Article Types EDA  - What Article Type Is Shared Mostly Frequently?


```python
article_types = [(x, cleaned_df.groupby(x)) for x in cleaned_df.columns if x.startswith('article')]
```

```python
fix, ax = plt.subplots(figsize=(20,9))
topic_data = [(item[0], item[1].Total_Shares.mean().values[1]) for item in article_types]
plt.barh([item[0] for item in topic_data], width = [item[1] for item in topic_data] )
plt.xticks(rotation = 90)
plt.xlabel('Mean Shares - μ ', fontsize='16', labelpad=20)
plt.ylabel('Article Types',  fontsize='16', labelpad=20)
plt.show()
```

Tagged articles with as either "how to articles" & "why post" on average received the most amount of shares, this was followed by infographics and list articles. 

Therefore the insight that we can gain from this is that people reading digital marketing topics share on average content that is educational (how to), visual (infographic) and easily digestible (list articles).

## Topic Type EDA - What Topic Is Shared Mostly Frequently?

```python
group_by_objects = []

for x in cleaned_df.columns:
    if x.startswith('Topic_'):
        group_by_objects.append((x , cleaned_df.groupby(x)))
```

```python
fix, ax = plt.subplots(figsize=(20,9))
topic_data = [(item[0], item[1].Total_Shares.mean().values[1]) for item in group_by_objects]
plt.bar([item[0] for item in topic_data], height = [item[1] for item in topic_data])
plt.xticks(rotation = 45)
plt.xlabel('Topics', fontsize='16', labelpad=20)
plt.ylabel('Mean Shares - μ',  fontsize='16', labelpad=20)
plt.show()
```

Search Engine Marketing was the most shared topic which was closely followed by Growth Marketing & Social Media Marketing. 

The mean shares was considerably lower for website design and display advertising and thus I would recommend creating content for the previously mentioned topics.

## Machine Learning  #### 

```python
cleaned_df.drop(axis=1, columns = ['Url', 'Published_Date', 'Setences_Text'], inplace=True)
```

```python
#1 Initialize The Scaler

scaler = StandardScaler()
X = cleaned_df
y = cleaned_df.pop('Total_Shares')
```

```python
#2 Setup A Train Test Split

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size = 0.2, random_state= 3)
```

```python
#3 Performing TFID Vectorizing Upon The Text Data

vectorizer = TfidfVectorizer(stop_words = 'english'
                            ,min_df=3, 
                             ngram_range=(1, 4),
                             max_features = 1000,
                            )

#Ensuring That All Of The Rows Are String Type

X_train['Article_Text'] = X_train['Article_Text'].astype(str)
X_test['Article_Text'] = X_test['Article_Text'].astype(str)

X2_train_titles = vectorizer.fit_transform(X_train['Article_Text'])
X2_test_titles = vectorizer.transform(X_test['Article_Text'])
```

```python
# Drop The Additional Text Columns

X_train.drop(axis=1, columns = ['Article_Text', 'Meta_Description', 'Title_Text'], inplace= True)
X_test.drop(axis=1, columns = ['Article_Text', 'Meta_Description', 'Title_Text'], inplace= True)
```

```python
for y in X_train.columns:
    if cleaned_df[y].dtype == 'object':
        print(y)
```

```python
#4 Apply Standardized Scaling To The Original Matrix

numerical_columns = [item for item in cleaned_df.columns if cleaned_df[item].dtype != 'object']
text_columns = [item for item in cleaned_df.columns if cleaned_df[item].dtype == 'object']

X_train_std = scaler.fit_transform(X_train.loc[:,numerical_columns])
X_test_std = scaler.transform(X_test.loc[:,numerical_columns])
```

```python
X_train_num = pd.DataFrame(X_train_std, columns = numerical_columns, index = X_train.index)
X_test_num = pd.DataFrame(X_test_std, columns = numerical_columns, index = X_test.index)
```

```python
#5 Combine The NLP Text Data With The Numerical Standardized Data

X_Train_Sparse = sparse.csr_matrix(X_train.values)
X_Test_Sparse = sparse.csr_matrix(X_test.values)

X2_train = sparse.hstack((X2_train_titles, X_Train_Sparse))
X2_test = sparse.hstack((X2_test_titles, X_Test_Sparse))
```
